```python
#Applying Linear Model on the new_data_set1

## separating the data and labels
X=new_data_set1[['PH-121','HS-101','CY-105','HS-105/12','MT-111','CS-105','CS-106','EL-102','EE-119','ME-107','CS-107']]
Y=new_data_set1['CGPA']

X_train,X_test,Y_train,Y_test =train_test_split(X, Y,test_size=0.2,random_state=0)
model1_linear=LinearRegression()
model1_linear.fit(X_train,Y_train)

```
